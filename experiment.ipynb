# comparer les algorithmes

```python


num_req = 5
total_obs = []
times = [[],[],[],[]]
rewards = [[],[],[],[]]
for i in range(10):
	eoscsp = generate_random_esop_instance(4, 3, num_req)
	num_req += 1
	
	total_obs.append(len(eoscsp.observations))

	start = time.time()
	schedule,_,reward = greedy_eoscsp_solver(eoscsp)
	end = time.time()
	times[0].append(end-start)
	rewards[0].append(reward)

	start = time.time()
	schedule,reward = s_dcop(eoscsp)
	end = time.time()
	times[1].append(end-start)
	rewards[1].append(reward)

	start = time.time()
	schedule,reward = psi_solver(eoscsp)
	end = time.time()
	times[2].append(end-start)
	rewards[2].append(reward)

	start = time.time()
	schedule,reward = ssi_solver(eoscsp)
	end = time.time()
	times[3].append(end-start)
	rewards[3].append(reward)

# plot the results
plt.figure()
plt.plot(total_obs,times[0],label='Greedy')
plt.plot(total_obs,times[1],label='DCOP')
plt.plot(total_obs,times[2],label='PSI')
plt.plot(total_obs,times[3],label='SSI')
plt.xlabel('Number of observations')
plt.ylabel('Time (s)')
plt.yscale('log')
plt.legend()
plt.show()

plt.figure()
plt.plot(total_obs,rewards[0],label='Greedy')
plt.plot(total_obs,rewards[1],label='DCOP')
plt.plot(total_obs,rewards[2],label='PSI')
plt.plot(total_obs,rewards[3],label='SSI')
plt.xlabel('Number of observations')
plt.ylabel('Reward')
plt.legend()
plt.show()



```
